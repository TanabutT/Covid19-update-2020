```python
from openpyxl import load_workbook
from datetime import date
import requests
from pprint import pprint

today = date.today().strftime('%Y%m%d')

wb = load_workbook(r"C:\Users\Ttaksinavongskul\Documents\template covid19.xlsx")
#put r to raw data 
ws = wb["Sheet1"]



r = requests.get("https://coronavirus-19-api.herokuapp.com/countries")

data = r.json()

for i in range(1,20):
        Colread = ws.cell(row=2, column=i).value
        print(i, Colread) # check index and column name
        
        if Colread == "country":
            indexcountry = i
        if Colread == "cases":
            indexcases = i                   
        if Colread == "todayCase":
            indextodayCase = i 
        if Colread == "deaths":
            indexdeaths = i
        if Colread == "todayDeaths":
            indextodayDeaths = i 
        if Colread == "recovered":
            indexrecovered = i
        if Colread == "active":
            indexactive = i
        if Colread == "critical":
            indexcritical = i
        if Colread == "casesPerOneMillion":
            indexcasesPerOneMillion = i
            
print(indexcountry) # check for loop   


for i in range(len(data)):
   # print(data[i]["country"])
   # print(data[i])
    datacountry = (data[i]["country"])
    datacases = (data[i]["cases"])
    datatodayCases = (data[i]["todayCases"])
    datadeaths = (data[i]["deaths"])       
    datatodayDeaths = (data[i]["todayDeaths"])
    datarecovered = (data[i]["recovered"])
    dataactive = (data[i]["active"])
    datacritical = (data[i]["critical"])
    datacasesPerOneMillion = (data[i]["casesPerOneMillion"])
    
    #GetData
    ws.cell(row=i+3, column=indexcountry).value = datacountry
    ws.cell(row=i+3, column=indexcases).value = datacases
    ws.cell(row=i+3, column=indextodayCase).value = datatodayCases
    ws.cell(row=i+3, column=indexdeaths).value = datadeaths
    ws.cell(row=i+3, column=indextodayDeaths).value = datatodayDeaths
    ws.cell(row=i+3, column=indexrecovered).value = datarecovered
    ws.cell(row=i+3, column=indexactive).value = dataactive
    ws.cell(row=i+3, column=indexcritical).value = datacritical
    ws.cell(row=i+3, column=indexcasesPerOneMillion).value = datacasesPerOneMillion
    
    
ws.cell(row=1, column=1).value = f"Covid Data {today}"
wb.save(r"C:\Users\Ttaksinavongskul\Documents\Covid19_{}.xlsx.".format(today))

    
    
```
